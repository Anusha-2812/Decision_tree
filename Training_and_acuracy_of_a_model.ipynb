```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/pima-indians-diabetes.data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# # x and y iloc=integerlocation, used because column names are also integers in the above dataset
x=df.iloc[:,:8]
y=df.iloc[:,-1]
```

```python
seed=7
test_size=0.30
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=test_size,random_state=seed)
```

```python
#train the model
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
predictions=[round(value) for value in y_pred]
print(predictions)
```

```python
accuracy=accuracy_score(y_test,predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/Water_Potability_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
x=df.iloc[:,:8]
y=df.iloc[:,-1]
```

```python
seed=7
test_size=0.30
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=test_size,random_state=seed)
```

```python
#train the model
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
predictions=[round(value) for value in y_pred]
print(predictions)
```

```python
accuracy=accuracy_score(y_test,predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
